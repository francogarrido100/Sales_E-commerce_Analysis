```python
import pandas as pd
df = pd.read_csv('sale report.csv')
df.head()

```

```python
df.groupby('Color')['Stock'].sum()
```

```python
df['Category'].value_counts()
```

```python
stock_por_color = df.groupby('Color')['Stock'].sum()
variantes_por_categoria = df['Category'].value_counts()
with pd.ExcelWriter('Reporte_Ventas_Limpio.xlsx') as writer:
    stock_por_color.to_excel(writer, sheet_name='Stock por Color')
    variantes_por_categoria.to_excel(writer, sheet_name='Variantes por Categoría')
```

```python
df['Category'].value_counts().head(5).plot(kind='bar', color='darkblue')
```
